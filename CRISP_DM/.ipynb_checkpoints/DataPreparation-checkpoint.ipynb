```python
# De forma similar, en el caso de baños..
# 0: No especificado

df["Baños"].unique()
```

```python
weird_banos_mask = ((df["Baños"] == 0) & (df["Precio"] > 1000000000))

df[weird_banos_mask].count()

# Hay 20 propiedades de más de 150 millones sin baños, esto es claramente un default del sistema.
# Para rellenar estos datos y que sean útiles haré algo un poco arbitrario.

# Primero haré un group-by por precio y hacer un promedio de baños. Para cada precio de los 20 que aparecen les asignaré el promedio
# de baños de esa categoría.

# Antes, eliminemos datos extremos en la categoría de precios...
```

```python
# Elimino propiedades mayores a 2,000,000,000 eso son 10 mil millones de pesos. 

df[df["Precio"] > 1.5e9]["ID de la propiedad"].count()
```

```python
# Sólo hay nueve, además, no es lo que estudiamos

df = df[df["Precio"] <= 1.5e9]
```

```python
num_bins = 10

temp = pd.DataFrame()
temp['ID de la propiedad'] = df['ID de la propiedad'].copy()

min_precio = df['Precio'].min()
max_precio = df['Precio'].max()

temp['rango_precio'] = pd.cut(df['Precio'], bins=num_bins, 
                              include_lowest=True, 
                              right=True).copy()

bins = np.linspace(min_precio, max_precio, num_bins+1)
temp['rango_precio'] = pd.cut(df['Precio'], bins=bins, include_lowest=True).copy()
temp['Baños'] = df['Baños'].copy()
temp['rango_precio'] = pd.IntervalIndex(temp['rango_precio'])
```

```python
# Excluimos a los datos afectados
temp = temp[temp["Baños"] != 0]
```

```python
# Agrupar por rango de precios y calcular promedio de Baños
promedio_baños = temp.groupby('rango_precio')['Baños'].mean().reset_index()

# Renombrar la columna para mayor claridad
promedio_baños.rename(columns={'Baños': 'promedio_baños'}, inplace=True)



promedio_baños
```

```python
interval_to_baños = pd.Series(promedio_baños['promedio_baños'].values, index=promedio_baños['rango_precio'])
```

```python
interval_to_baños
```

```python
precios = df[df["Baños"] == 0]["Precio"].copy()
```

```python
promedio_baños = precios.apply(lambda x: interval_to_baños[interval_to_baños.index.contains(x)].iloc[0])
```

```python
promedio_baños
```

```python
df.loc[df["Baños"] == 0, "Baños"] = promedio_baños
```

```python
# Función para encontrar el barrio asociado a las coordenadas exactas
def asignar_barrio(lat, lon, df_barrio):
    # Filtramos el DataFrame df_barrio para encontrar una coincidencia exacta de latitud y longitud
    match = df_barrio[(df_barrio['Latitud'] == lat) & (df_barrio['Longitud'] == lon)]
    
    if not match.empty:
        # Si encontramos una coincidencia, devolvemos el nombre del barrio
        return match['Barrio'].iloc[0]
    else:
        # Si no hay coincidencia, devolvemos None
        return None

# Aplicamos la función a cada fila de no_tiene_barrio para asignar la Localidad Principal
no_tiene_barrio['Localidad Principal'] = no_tiene_barrio.apply(
    lambda row: asignar_barrio(row['Latitud'], row['Longitud'], barrios), axis=1
)

# Asignamos el valor 'neighbourhood' a la columna 'Tipo de Localidad Principal'
no_tiene_barrio['Tipo de Localidad Principal'] = 'neighbourhood'

# Verificamos los primeros resultados
print(no_tiene_barrio[['ID de la propiedad', 'Latitud', 'Longitud', 'Localidad Principal', 'Tipo de Localidad Principal']].head())
```

### Código generador de barrios por coordenada 

```python
import requests
import json
import os
import time
from concurrent.futures import ThreadPoolExecutor, as_completed
from threading import Lock
from itertools import cycle
import pandas as pd

# -----------------------------
# Configuración
# -----------------------------
OVERPASS_MIRRORS = [
    "https://overpass-api.de/api/interpreter",
    "https://lz4.overpass-api.de/api/interpreter",
    "https://overpass.kumi.systems/api/interpreter",
    "https://overpass.openstreetmap.fr/api/interpreter"
]

mirror_cycle = cycle(OVERPASS_MIRRORS)
mirror_lock = Lock()
mirror_sleep = {mirror: 0 for mirror in OVERPASS_MIRRORS}  # timestamps para bloquear mirrors temporalmente
mirror_backoff = {mirror: 2.5 for mirror in OVERPASS_MIRRORS}  # tiempo de espera para cada mirror en caso de 429

RADIUS = 800
SLEEP_BETWEEN_QUERIES = 2.5  # pausa inicial entre queries
MAX_WORKERS = 5
BATCH_SIZE = 10  # Escribimos cada vez que tengamos 10 registros de lat, lon y barrio

progress_lock = Lock()
completed_queries = 0

# -----------------------------
# Funciones
# -----------------------------
def get_next_mirror():
    """Obtiene el siguiente mirror disponible, respetando los tiempos de espera por 429."""
    with mirror_lock:
        for _ in range(len(OVERPASS_MIRRORS)):
            mirror = next(mirror_cycle)
            if time.time() >= mirror_sleep[mirror]:
                return mirror
        # Si todos están bloqueados, esperar 1 segundo y reintentar
        time.sleep(1)
        return get_next_mirror()

def save_batch(batch, output_file):
    """Guarda un batch de resultados en JSON (append)."""
    if os.path.exists(output_file) and os.path.getsize(output_file) > 0:
        with open(output_file, "r", encoding="utf-8") as f:
            all_results = json.load(f)
    else:
        all_results = []

    all_results.extend(batch)
    with open(output_file, "w", encoding="utf-8") as f:
        json.dump(all_results, f, ensure_ascii=False, indent=2)

def query_overpass(lat, lon):
    global completed_queries
    overpass_url = get_next_mirror()
    query = f"""
    [out:json][timeout:90];
    (
      node(around:{RADIUS},{lat},{lon})[place=neighbourhood];
      way(around:{RADIUS},{lat},{lon})[place=neighbourhood];
      relation(around:{RADIUS},{lat},{lon})[place=neighbourhood];
    );
    out center;
    """
    try:
        response = requests.get(overpass_url, params={"data": query}, timeout=120)
        response.raise_for_status()
        result_data = response.json()

        # Si hay un resultado, extraemos el barrio (nombre del lugar)
        if result_data['elements']:
            barrio = result_data['elements'][0].get('tags', {}).get('name', 'Desconocido')
        else:
            barrio = 'Desconocido'
        
        result = {"Latitud": lat, "Longitud": lon, "Barrio": barrio}
        mirror_backoff[overpass_url] = 2.5  # Reset the backoff time after a successful request

    except requests.exceptions.HTTPError as e:
        print(f"HTTP error {e.response.status_code} en {overpass_url} para {lat},{lon}")
        if e.response.status_code == 429:
            # Congelar el mirror por más tiempo si sigue fallando con 429
            mirror_backoff[overpass_url] *= 2  # Duplicar tiempo de espera en caso de 429
            mirror_sleep[overpass_url] = time.time() + mirror_backoff[overpass_url]
        result = None
    except requests.exceptions.RequestException as e:
        print(f"Error de solicitud en {overpass_url} para {lat},{lon}: {e}")
        result = None

    with progress_lock:
        completed_queries += 1
        print(f"Queries completadas: {completed_queries}")

    return result

def process_coordinates_parallel(df, output_file="barrios.json"):
    global completed_queries
    completed_queries = 0

    if os.path.exists(output_file) and os.path.getsize(output_file) > 0:
        with open(output_file, "r", encoding="utf-8") as f:
            results = json.load(f)
    else:
        results = []

    total_queries = len(df)
    print(f"Total de queries a procesar: {total_queries}")

    batch_results = []
    tasks = []
    with ThreadPoolExecutor(max_workers=MAX_WORKERS) as executor:
        for _, row in df.iterrows():
            lat, lon = row["Latitud"], row["Longitud"]
            tasks.append(executor.submit(query_overpass, lat, lon))

        for future in as_completed(tasks):
            result = future.result()
            if result:
                batch_results.append(result)

            # Guardar cada vez que tengamos 10 registros en el batch
            if len(batch_results) >= BATCH_SIZE:
                save_batch(batch_results, output_file)
                batch_results = []  # Limpiar el batch

            time.sleep(SLEEP_BETWEEN_QUERIES)

    # Guardar lo que quede en el batch
    if batch_results:
        save_batch(batch_results, output_file)

    print(f"Finalizado. {completed_queries}/{total_queries} queries completadas. Resultados guardados en {output_file}.")
```

```python
# Crear una copia temporal de la columna para no modificar df
met_cuad = df['Metros Cuadrados']

# Calcular min y max de metros cuadrados
min_m2 = met_cuad.min()
max_m2 = met_cuad.max()

# Definir número de grupos
n_grupos = 3

# Crear bins (intervalos)
bins = np.linspace(min_m2, max_m2, n_grupos + 1)

# Crear etiquetas con formato limpio "a–b"
labels = [f"{int(bins[i])}–{int(bins[i+1])}" for i in range(len(bins)-1)]

# Crear serie de grupo sin modificar df
grupo_m2 = pd.cut(
    met_cuad,
    bins=bins,
    labels=labels,
    include_lowest=True
)

# Calcular distribución porcentual por grupo de metros cuadrados
dist_porcentual = (
    df.assign(**{'Grupo Metros Cuadrados': grupo_m2})  # se crea temporalmente
    .groupby(['Grupo Metros Cuadrados', 'Antigüedad (Grupo)'])
    .size()
    .groupby(level=0)
    .apply(lambda x: 100 * x / x.sum())
    .unstack(fill_value=0)
)

# Gráfico apilado porcentual
ax = dist_porcentual.plot(
    kind='bar',
    stacked=True,
    figsize=(9, 6),
    colormap='viridis'
)

plt.title('Distribución porcentual de Antigüedad por grupo de Metros Cuadrados')
plt.xlabel('Grupo de Metros Cuadrados')
plt.ylabel('Porcentaje (%)')
plt.xticks(rotation=45)
plt.legend(title='Antigüedad (Grupo)', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

### ¿Cuál enfoque elegir?
Tenemos dos opciones:

A) Elegir precio promedio de las casas del barrio

B) Elegir precio promedio del metro cuadrado del barrio


Eligiré la segunda opción. Esto porque es una medida bien conocida en finca raiz, contrario a la primera que se puede deducir de la segunda.

### ¿Qué se hará con aquellas propiedades cuyo barrio no tenga esta cantidad mínima de propiedades?
Se le asignará la mediana de los valores anteriormente mostrados.

### Uso en el despliegue
Siempre hay que pensar en cómo se usará el modelo. Este dato se autorellenará, de ser posible, con los datos de un geo reverse-lookup usando la ubicación suministrada.


Abajo se implementa este enfoque


```python
dm = df.copy()
dm["precio_metro_cuadrado"] = (dm["Precio"] / dm["Metros Cuadrados"])

mediana_precio_metro_cuadrado = dm.groupby(["Ciudad", "Estrato"]).median("precio_metro_cuadrado")["precio_metro_cuadrado"]
```

```python
mediana_precio_metro_cuadrado.loc["Bello", 1]
```

```python
def return_m2_price(index):
    if index in precio_por_metro_cuadrado: 
        return precio_por_metro_cuadrado.loc[index]
    else:
        return None

def obtener_precio_por_metro(row):
    # Verificar si la combinación de Ciudad y Estrato está en el índice de mediana_precio_metro_cuadrado
    try:
        return mediana_precio_metro_cuadrado.loc[row["Ciudad"], row["Estrato"]]
    except KeyError:
        # Si no se encuentra la clave, se puede devolver un valor por defecto (ej. None o algún valor específico)
        return mediana_precio_metro_cuadrado.median()  # O cualquier valor que prefieras como fallback


# Ahora aplicamos esta función a las filas con NaN en "Precio por Metro Cuadrado Sector"


dm["Precio por Metro Cuadrado Sector"] = dm["Localidad Principal"].apply(return_m2_price)

dm.loc[dm["Precio por Metro Cuadrado Sector"].isnull(), "Precio por Metro Cuadrado Sector"] = dm.apply(lambda row: obtener_precio_por_metro(row), axis=1)
```

```python
dm["Precio por Metro Cuadrado Sector"].isnull().sum()
```

```python

# Se eliminan las columnas que no aportan nada al modelado.
dm = dm.drop(['ID de la propiedad', 'URL', 'Dirección', "Localidad Principal"], axis = 1)

dm = dm.dropna(subset=['Antigüedad (Grupo)']) # Linea Provisional (Problema: Se eliminan 884 datos, es posible que haya manera de llenarlos)
dm = dm.dropna(subset = ['Metros Cuadrados']) 
dm = dm[(dm['Estrato'] < 7) & (dm['Estrato'] > 0)]
dm = dm[dm['Parqueaderos'] >=0]
dummies = pd.get_dummies(
    dm[["Tipo de Inmueble", "Ciudad"]],
    prefix=["tipo", "ciudad"],
    
)


# Información Costo Promedio del Metro Cuadrado según Barrio (Con más de 40 propiedades):


# --- Añadirlas al DataFrame original ---
dm = pd.concat([dm, dummies], axis=1)

# 🧹 --- Eliminar las columnas originales ---
dm = dm.drop(columns=["Tipo de Inmueble", "Ciudad"])
```

```python
dm['Estrato'] = dm['Estrato'].round()
dm['Antigüedad (Grupo)'] = dm['Antigüedad (Grupo)'].round()
```

```python
dm.drop(columns=["precio_metro_cuadrado"], inplace=True)
```
