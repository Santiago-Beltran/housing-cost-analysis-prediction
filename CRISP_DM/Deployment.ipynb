```python
import pandas as pd
from datetime import datetime
import pickle as pk

# Load the model
loaded_model = pk.load(open("modelo_final_escogido.pickle", "rb"))
```

```python
mediana_precio_metro_cuadrado = pd.read_json("medianas.json", orient="table")

ciudades = {
    1: "Bello",
    2: "Envigado",
    3: "Itaguí",
    4: "La Estrella",
    5: "Medellín",
    6: "Sabaneta"
}

def obtener_precio_por_metro(ciudad, estrato):
    # Verificar si la combinación de Ciudad y Estrato está en el índice de mediana_precio_metro_cuadrado
    try:
        return mediana_precio_metro_cuadrado.loc[ciudad, estrato].iloc[0]
    except KeyError:
        # Si no se encuentra la clave, se puede devolver un valor por defecto (ej. None o algún valor específico)
        return mediana_precio_metro_cuadrado.median().iloc[0] # O cualquier valor que prefieras como fallback
```

```python
# Rellenar características.

m2 = float(input("Metros Cuadrados: "))
```

```python
Ubicacion = input("Ubicación de Google Maps: ") # 6.175082, -75.576859
```

```python
Latitud, Longitud = Ubicacion.split(',')

Latitud = float(Latitud)
Longitud = float(Longitud)
```

```python
Antiguedad = int(input("Antigüedad (Grupo del 1 al 5): "))
```

```python
Estrato = int(input("Estrato (1 a 6): "))
```

```python
Num_dormitorios = int(input('Dormitorios (bedrooms): '))
Banos = int(input("Baños: "))
Num_parqueaderos = int(input("Num parqueaderos: "))
```

```python
Fecha_creacion_oferta = input("Fecha de Creacion de Oferta YYYY-MM-DD: ")
Fecha_actualizacion_oferta = input("Fecha de Creacion de Oferta YYYY-MM-DD: ")

Fecha_creacion_oferta = datetime.strptime(Fecha_creacion_oferta, "%Y-%m-%d")
Fecha_actualizacion_oferta = datetime.strptime(Fecha_actualizacion_oferta, "%Y-%m-%d")
```

```python
decision_tipo_apto = int(input("[1]: Casa o [2]: Apto: "))
print()

decision_ciudad = int(input(" CIUDAD: \n[1]: Bello \n[2]: Envigado \n[3]: Itagüí \n[4]: La Estrella \n[5]: Medellín \n[6]: Sabaneta \n > "))
```

```python
decision_precio_metro_cuadrado_promedio = int(input("Conoce usted el precio del metro cuadrado promedio de su zona? \n[1] Sí\n[2] No"))

if decision_precio_metro_cuadrado_promedio == 1:
    precio_metro_cuadrado = float(input("Introduzca este valor $/m2: "))
else:
    precio_metro_cuadrado = obtener_precio_por_metro(ciudades[decision_ciudad], Estrato)
    
```

```python
# --- Inicializar DataFrame con todas las columnas en 0/NaN ---
columnas = ['Metros Cuadrados', 'Estrato', 'Latitud', 'Longitud',
            'Antigüedad (Grupo)', 'Dormitorios (bedrooms)', 'Baños', 'Parqueaderos',
            'Fecha de Creacion de Oferta', 'Fecha de Actualización de Oferta', 
            "Precio por Metro Cuadrado Sector",
            'tipo_Apartamento', 'tipo_Casa', 'ciudad_Bello',
            'ciudad_Envigado', 'ciudad_Itaguí', 'ciudad_La estrella', 
            'ciudad_Medellín', 'ciudad_Sabaneta']

df = pd.DataFrame([[0]*len(columnas)], columns=columnas)

# --- Llenar valores numéricos / fechas ---
df['Metros Cuadrados'] = m2
df['Estrato'] = Estrato
df['Latitud'] = Latitud
df['Longitud'] = Longitud
df['Antigüedad (Grupo)'] = Antiguedad
df['Dormitorios (bedrooms)'] = Num_dormitorios
df['Baños'] = Banos
df['Parqueaderos'] = Num_parqueaderos
df['Fecha de Creacion de Oferta'] = int(Fecha_creacion_oferta.timestamp())
df['Fecha de Creacion de Oferta'] = int(Fecha_creacion_oferta.timestamp())


df['Fecha de Actualización de Oferta'] = int(Fecha_actualizacion_oferta.timestamp()) 

df["Precio por Metro Cuadrado Sector"] = precio_metro_cuadrado

# --- Llenar columnas one-hot según selección ---
if decision_tipo_apto == 1:
    df['tipo_Casa'] = 1
    df['tipo_Apartamento'] = 0
else:
    df['tipo_Casa'] = 0
    df['tipo_Apartamento'] = 1

ciudades = {
    1: 'ciudad_Bello',
    2: 'ciudad_Envigado',
    3: 'ciudad_Itaguí',
    4: 'ciudad_La estrella',
    5: 'ciudad_Medellín',
    6: 'ciudad_Sabaneta'
}

for key in ciudades:
    if key != decision_ciudad:
        df[ciudades[key]] = 0
    else:
        df[ciudades[key]] = 1
```

```python
predictions = loaded_model.predict(df)
```

```python
print(f"Precio Sugerido: ${predictions[0].round():,} COP")

df
```
